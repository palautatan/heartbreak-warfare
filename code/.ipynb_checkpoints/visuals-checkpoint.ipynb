# Logistic regression.

```r
training_set <- merge(train_values, train_labels, by="patient_id")
head(training_set)
```

```r
training_set %>% group_by(heart_disease_present) %>% summarize(mean_rbp=mean(resting_blood_pressure),
                                                               mean_age=mean(age),
                                                               mean_hr=mean(max_heart_rate_achieved),
                                                               mean_oldpeak=mean(oldpeak_eq_st_depression),
                                                               mean_choles=mean(serum_cholesterol_mg_per_dl))
```

```r
logit <- glm(heart_disease_present ~ resting_blood_pressure + age + max_heart_rate_achieved + oldpeak_eq_st_depression + serum_cholesterol_mg_per_dl, data=training_set, family="binomial")
```

```r
predictions <- round(predict(logit, test_values, type="response"))
```

```r
head(predictions)
```

```r
submission <- data.frame(cbind(test_values$patient_id, predict(logit, test_values, type="response")))
names(submission) <- c("patient_id", "heart_disease_present")
head(submission)
```

```r
write.csv(submission, "../results/submission_test.csv", row.names=FALSE)
```
